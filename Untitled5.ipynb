1. Sum of Even Numbers

Write a function that calculates the sum of all even numbers from 1 to a

given number.

```python
x=int(input())
sum=0
for i in range(x+1):
    if i%2==0:
        sum+=i
print(sum)
```

2. Prime Number Checker

Create a function that checks if a given number is prime.

```python
sum=0
x=int(input())
for i in range(1,x+1):
  if x%i==0:
    sum+=1
if sum ==2:
  print(f"the number {x} is a prime number")

```

3. Factorial Calculator

Write a function that calculates the factorial of a number using a loop.

```python
x=int(input())
m=1
while x:
  m*=x
  x-=1
print(m)
```

4. Fibonacci Sequence

Create a function that generates the Fibonacci sequence up to a given

number.

```python
x=int(input())
ls=[]
ls.append(0)
ls.append(1)
for i in range(2,x+1):
  if ls[i-2]+ls[i-1]>x:
    break
  ls.append(ls[i-2]+ls[i-1])

print(ls)
```

5. Multiplication Table

Write a function that prints the multiplication table of a number.


```python
x=int(input())
for i in range(1,11):
  print(f"{x} X {i} = {x*i}")
```

6. Reverse a String

Write a function that reverses a string.

```python
x=input()
y=len(x)
for i in range(y):
  print(x[y-1],end="")
  y-=1
```

8. Sum of Digits

Write a function to calculate the sum of digits of a number

```python
x=int(input())
sum=0
while x:
  sum+=int(x%10)
  x=int(x/10)
print(sum)
```

9. Count Vowels

Create a function that counts the number of vowels in a string.

```python
x=input()
sum=0
for i in x:
  if i in "aeiouAEIOU":
    sum+=1
print(sum)
```
